Для этого были построены сводные таблицы в разбивке по категориям, графическая визуализация, чтобы наглядно продемострировать наличие или отсутствие зависимости между показателями.

В результате анализа исходные данные были очищены от пропущенных и аномальных значений, что сделало данные более надежными и годными для анализа.

Анализ данных показал, что стоимость квартир больше всего зависит от:
- Места расположения - средняя стоимость одного квадратного метра в Санкт-Петербурге намного выше среди 10 населённых пунктов с наибольшим числом объявлений. Разница в стоимости почти в два раза.
- Общей и жилой площади объекта - чем больше площадь объекта, тем больше его стоимость.
- Количества комнат - объекты с бОльшим количеством комнат стоят дороже
- Типа этажа. Стоимость квартир на первом этаже значительно ниже, чем на средних и последних этажах
- Даты размещения - день недели, месяц, год размещения не оказывают влияния на стоимость квартир.

Была проанализирована средняя стоимость квартир в городе Санкт-Петербурге в зависимости от расстояния до центра города. Анализ показал, что наиболее дорогие квартиры находятся на расстоянии до 8 км от центра города
